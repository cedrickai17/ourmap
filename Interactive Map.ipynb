```python
import folium
import base64
import os
import pandas as pd
from github import Github
```

```python
sheet_id = '1j1T54UNgcUnM80YTbwn158PfrOuGOk8C5KBiAHTaHik'
sheet_name = 'Sheet1'
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"

df_google_sheet = pd.read_csv(url)
df_google_sheet
```

```python
m = folium.Map(location=[1.349551, 103.812823], tiles='CartoDB positron',zoom_start=12)
```

```python
html = folium.Html (
    f"""
    <html>
    <h3>Our Long Awaited Cafe</h3>
    <p>the ice cream was good!</p>
    <img src= 'beauty_1676803468093.JPG' width=100px>
    </html>
""", script=True
)
```

```python
def create_marker(df):
    for i in range(0,len(df)):
        #not using this iframe for now. not sure what's its for
        #iframe = folium.IFrame(html=html,width=400,height=400)
        marker_popup = folium.Popup(html=html, max_width=500)
        marker_tooltip = folium.Tooltip(text=df.iloc[i]['Address'],sticky=False)
        marker_icon = folium.Icon(color="blue",icon="star", prefix='fa')
        #create marker from here
        folium.Marker([df.iloc[i]['Lat'],df.iloc[i]['Lon']], popup=marker_popup, tooltip=marker_tooltip, icon=marker_icon).add_to(m)
```

```python
create_marker(df_google_sheet)
```

```python
m.save('index.html')
```
