```python
m = folium.Map(location=[1.349551, 103.812823], tiles='CartoDB positron',zoom_start=12)
m
```

```python
def create_marker(df):
    for i in range(0,len(df)):
        iframe = folium.IFrame(html=html,width=400,height=400)
        marker_popup = folium.Popup(html=html, max_width=500)
        marker_tooltip = folium.Tooltip(text=df.iloc[i]['Address'],sticky=False)
        marker_icon = folium.Icon(color="blue",icon="star", prefix='fa')
        #create marker from here
        folium.Marker([df.iloc[i]['Lat'],df.iloc[i]['Lon']], popup=marker_popup, tooltip=marker_tooltip, icon=marker_icon).add_to(m)
    


```

```python
create_marker(df)
```
